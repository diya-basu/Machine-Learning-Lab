#There are few missing values like ["n/a", "na",  "--"] Apply some technique to replace these values with NaN.

```python
dt=pd.DataFrame(data)
dt=pd.DataFrame(data,columns=["SQ_FT"])
dt['SQ_FT']=dt['SQ_FT'].replace(['n/a', 'na', '--'], 'NaN')
```
